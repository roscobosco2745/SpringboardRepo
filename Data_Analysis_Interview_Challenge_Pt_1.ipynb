1) I would say the key measure of this experiment would be how much money is reimbursed to the drivers.  The fact that money is being reimbursed indicates that a driver left one city to go to the other.  I could not simply use the total number of tolls as the metric because the driver partners would still take tolls from people travelling within their perspective cities.  In addition, I would only consider the reimbursed fares for the weekdays since there was already activity between the two cities during the weekend.

2) I would issue all the driver partners a special tag and then add a sensor at the toll that scans those tags.  The sensor would only be active on the weekdays.  Before actually implementing the experiment I would take note of the average number of hits the sensor would take on the weekdays over a period of 4 weeks.  This will allow for the calculation of the average number of travels between the cities.  Then I would turn record the sensors after telling the drivers they would be reimbursed for all tolls.
   I would look at two statistical measures: the average number of detected sensor triggerings per week (during the weekdays) and 2) The variation of the number of triggerings from the mean.  I would be looking to see if the mean increases significantly over the weeks following the initial readings, and I would watch to see if the variance is low, which would indicate if there is another force affecting the drivers crossing over into the other city.  This would help me recommend if the reimbursement incentive is a good idea and if there are conditions that make it not such a good idea.
